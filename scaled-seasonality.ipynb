## Seasonality Indices Calculation

```python
import pandas as pd

#WE ADD ALL THE DATA FROM YEARS INTO ONE BIG DATAFRAME
#UPDATE YOUR FILE PATH ACCORDINGLY
df = pd.read_excel(r'D:\Projects\ntu-vish\Yearly Energy Demand Data\2004.xlsx', index_col=0)
year = 2004
for i in range(1,19):
    temp_df = pd.read_excel("D:\\Projects\\ntu-vish\\Yearly Energy Demand Data\\" + str(year + i) + '.xlsx', index_col=0)
    df = df.merge(temp_df, left_index=True, right_index=True)
df.columns = pd.to_datetime(df.columns, dayfirst=True) # converting into datetime format
df.drop(df.columns[-1], axis=1, inplace=True) # removing Jan 01 2023 column
hourlyall = df.copy() # DataFrame with half-hourly values
```

* ### S.I. for the Month  
**Previous work**- applied the same formula, calculated weekdays and weekends separately  
**Current work**- same formula, weekdays and weekends not separated, only one value for a particular month, overall S.I. for a month 

```python
df = df.sum() # daily demand (summing up hourly demands for a day), also converting it into a pandas Series
dailyall = df.copy() # storing daily demand data
```

```python
dfdict = {}
for year in df.index.year.unique():
    dfdict[year] = []
    yearly_temp = df[df.index.year == year] # temporary yearly series
    yearly_avg = yearly_temp.mean() # average daily demand for the year
    for month in df.index.month_name().unique(): # [January,...,December]
        monthly_temp = yearly_temp[yearly_temp.index.month_name() == month] # temporay monthly series
        monthly_avg = monthly_temp.mean() # average daily demand for the month and year
        monthlysi = monthly_avg/yearly_avg
        dfdict[year].append(monthlysi)
monthlydf = pd.DataFrame(dfdict, index=df.index.month_name().unique()) # DataFrame holding S.I. for the month

#monthlydf.to_excel('S.I. for the Month [01.08].xlsx')
monthlydf
```

### Correction in the formula - 2004 in denominator instead of 2003

```python
overall_monthly = monthlydf.sum(axis = 1) / (2022-2004+1) # Overall S.I. for the Month
# with pd.ExcelWriter('S.I. for the Month [01.08].xlsx', mode='a', engine='openpyxl') as writer:
#     overall_monthly.to_excel(writer, sheet_name='Overall S.I. for the Month')
overall_monthly
```

* ### S.I. for the Day of the Week  
**Previous work**- no previous work   
**Current work**- formula as suggested, overall S.I. for day of the week

```python
for year in dailyall.index.year.unique():
    dfdict = {}
    for month in dailyall.index.month_name().unique():
        dfdict[month] = []
        for day in dailyall.index.day_name().unique():
            daily_temp = dailyall[dailyall.index.strftime("%Y%B%A") == str(year)+month+day]
            daily_avg = daily_temp.mean() / overall_monthly.loc[month] # average day of the week demand for a given month and year (divided by S.I. of the month)
            dfdict[month].append(daily_avg)
        dfdict[month] = dfdict[month] / (sum(dfdict[month])/len(dfdict[month])) # S.I. for the day of the week (for particular month and year)
    year_temp = pd.DataFrame(dfdict, index=dailyall.index.day_name().unique())
    # try:
    #     with pd.ExcelWriter('S.I. Day of the Week [01.08].xlsx', mode='a', engine='openpyxl') as writer:
    #         year_temp.to_excel(writer, sheet_name=str(year))
    # except:
    #     year_temp.to_excel('S.I. Day of the Week [01.08].xlsx', sheet_name=str(year))
```

```python
overall_weekly = pd.read_excel('S.I. Day of the Week [01.08].xlsx', index_col=0, sheet_name=str(2004)).sum(axis = 1)
for year in range(2005,2023):
    overall_weekly += pd.read_excel('S.I. Day of the Week [01.08].xlsx', index_col=0, sheet_name=str(year)).sum(axis = 1)
overall_weekly = overall_weekly / (19*12) # (19*12)- number of year and month combinations

# with pd.ExcelWriter('S.I. Day of the Week [01.08].xlsx', mode='a', engine='openpyxl') as writer:
#     overall_weekly.to_excel(writer, sheet_name='Overall S.I. Day of the Week')
overall_weekly
```

* ### S.I. for the Hour of the Day
**Previous work**- calculated weekdays, weekends, Mon-Fri separately, calculated both hourly and half-hourly, no other SIs were considered  
**Current work**- other SIs considered, all weekdays separately calculated, only hourly calculated, overall S.I. for hour of the day

```python
# changing data from half-hourly to hourly by adding consecutive half-hour data
drop_list = []
for i in range(48):
    if i%2 == 1:
        hourlyall.iloc[i] = (hourlyall.iloc[i] + hourlyall.iloc[i-1])
    else:
        drop_list.append(hourlyall.index[i])
hourlyall.drop(drop_list,inplace=True)
hourlyall.head()
```

```python
dfdict = {}
for year in hourlyall.columns.year.unique():
    for month in hourlyall.columns.month_name().unique():
        for day in hourlyall.columns.day_name().unique():
            dfdict[(year,month,day)] = []
            for hour in hourlyall.index:
                # average hourly demand for a gicen day of the week, month and year (divided by S.I. of the month and day of the week)
                hourly_avg = hourlyall[hourlyall.columns[hourlyall.columns.strftime("%Y%B%A") == str(year)+month+day]].loc[hour].mean() / (overall_monthly.loc[month]*overall_weekly.loc[day])
                dfdict[(year,month,day)].append(hourly_avg)
            # S.I. for the hour of the day (for particular day of the week, month and year)
            dfdict[(year,month,day)] = dfdict[(year,month,day)] / (sum(dfdict[(year,month,day)])/len(dfdict[(year,month,day)]))
hourlydf = pd.DataFrame(dfdict, index=hourlyall.index)
```
